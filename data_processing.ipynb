```python
import pandas as pd
import os 
import glob
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl
import datetime 

def process_excel_files(folder_path, new_path, path):
    # Ensure the new directories exist
    os.makedirs(new_path, exist_ok=True)
    os.makedirs(path, exist_ok=True)
    
    # Reading all the Excel files in the folder
    file_names = glob.glob(os.path.join(folder_path, "*.xls"))  # Adjust file extension if necessary

    # Iterate through each file
    for f in file_names:
        # Read Excel file into a DataFrame, skipping the first 4 rows
        df = pd.read_excel(f, skiprows=4)
        
        # Extract the base name of the file without extension and with a unique identifier
        base_name = os.path.splitext(os.path.basename(f))[0] if os.path.basename(f) else ''  # Get base name without extension
        base_name = base_name.split()[0] + '_' + str(file_names.index(f)) if base_name else 'file_' + str(file_names.index(f))  # Append unique index
        
        # Save it as a .csv in the new directory
        df.to_csv(os.path.join(path, f'{base_name}.csv'), index=False)

# Example usage:
# Define paths
folder_path = '/Users/jud-05/Desktop/High Court/4Q High Court'
new_path = '/Users/jud-05/Desktop/High Court/4Q combined'
path = '/Users/jud-05/Desktop/High Court/4Q combined/dcrt_4Q'

process_excel_files(folder_path, new_path, path)

# Get all the CSV files in the directory
all_files = glob.glob(path + "/*.csv")

all_files

# Combine all the CSV files into a single DataFrame and insert a column with the name of the file
li = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    df['file_name'] = os.path.basename(filename)
    li.append(df)
       
combined_csv = pd.concat(li, axis=0, ignore_index=True)


#save the combined file as a csv
combined_csv.to_csv(f'{path}/raw data/HC_combined.csv', index=False)

```
